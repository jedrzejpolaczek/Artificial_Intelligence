# Linear Regression 101

The Logistic Regression Algorithm basically goes like this:
* take your data,
* pick a random model (so basically weights and biases),
* calculate the error,
* minimize the error, and obtain a better model,
* profit!

So quick recap before we go further.
#### Recaption:
* multi class error function:
    * using cross-entropy: $-\frac{1}{n}{\sum_{i=1}^n{\sum_{j=1}^m y_{ij} ln(\hat y_{ij})}}$
    * squared errors (SSE): $E = \frac{1}{2}(y -\hat y)^2$
* error function (using cross-entropy): $-\frac{1}{n}{\sum_{i=1}^ny_iln(\hat y_i)} + (1 - y_i)(ln(1 -\hat y_i))$
* error value: $-(yln(\hat y) + (1 - y)(ln(1 -\hat y)))$
* expected output: $y_i = 0$ or $1$
* predicted output: probability $\hat y_i = \sigma(h_j)$
* avtivation function (may be):  
    * sigmoid (popular) $\sigma(h_i) = \frac{1}{1 + e^{-h_j}}$
    * softmax $\frac{e^{h_j}}{\sum_{i=1}^n e^{h_1}+...+e^{h_n}}$)
    * other.
* Linear function scores: $h_j = \sum_iW_ji\cdot x_i + b_j$ 
* Input: $X = (x_1, ..., x_n)$
* Model:
    * Weights: $W = (W_1, \dots, W_n)$ and $W_j = (w_j1, \dots, w_jn)$
    * Bias: $b = (b_1, \dots, b_n)$
* Sign for learning rate is $\alpha$

Let's get it all together now!

$E(W,b) = -\frac{1}{n}{\sum_{i=1}^n(1 - y_i)(ln(1 - \sigma(Wx^{(i)}+b)) - y_iln(\sigma(Wx^{(i)}+b)}$

Reminder! For multi class error function looks like that:

$E(W,b) = -\frac{1}{m}{\sum_{i=1}^n{\sum_{j=1}^m y_{ij} ln(\sigma(Wx^{(i)}+b))}}$

## Minimize Error Function
Ok so after recap we remind yourself the basics. We know how to calculate the error for each input and for the whole set of inputs. What next?
Obviously the goal is to lower the value given by $E(W,b) = -\frac{1}{m}{\sum_{i=1}^n{\sum_{j=1}^m y_{ij} ln(\sigma(Wx^{(i)}+b))}}$. We can use many different methods to do it. Mostly common is gradient descent. Let's move to it, shall we?

## Linear Regression Algorithm
1. Start with random weights: $w_1, \dots, w_n, b$ 
    * note: it will give us line that seperate point and probability function given by sigmoid: $\sigma(Wx + b)$.
2. Calculate error value for every point $x_i$ by using $E = -(1 - y)(ln(1 -\hat y)) - yln(\hat y)$
    * note: after substitution $E(W,b) = -\frac{1}{n}{\sum_{i=1}^n(1 - y_i)(ln(1 - \sigma(Wx^{(i)}+b)) - y_iln(\sigma(Wx^{(i)}+b)}$
    * note: error will be small for correctly classified points and high for incorrectly classified points.
3. For evry point $(x_1, \dots, x_n)$:
    * 3.1 For $i = 1, \dots, n$:
        * 3.1.1 Update $w_i$ by $w_i' \leftarrow w_i -\alpha\frac{\partial E}{\partial w_i}$
            * note: instead of $w_i' \leftarrow w_i -\alpha\frac{\partial E}{\partial w_i}$ we can use $w_i' \leftarrow w_i + \alpha (y - \hat{y}) x_i$
        * 3.1.2 Update $b$ by $b' \leftarrow b - \alpha\frac{\partial E}{\partial b}$
            * note: instead of $b' \leftarrow b - \alpha\frac{\partial E}{\partial b}$ we can use $b' \leftarrow b + \alpha (y - \hat{y})$
    * reminder: $\alpha$ is learning rate.
4. Repeat process until error is small
    * note: one pass of the process is called an epoch
    * note: in practice, we will repeat the process a fixed number of epochs

Example of the implementation below:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Some helper functions for plotting and drawing lines
def plot_points(X, y):
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'blue', edgecolor = 'k')
    plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'red', edgecolor = 'k')

data = pd.read_csv('gradient_descent_data.csv', header=None)
X = np.array(data[[0,1]])
y = np.array(data[2])
plot_points(X,y)
plt.show()
```

#### Implementation of function we talk about

Sigmoid activation function:

$\sigma(𝑥)=\frac{1}{1+e^{-x}}$

```python
# Activation (sigmoid) function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

Output (prediction) formula:

$\hat y = \sigma(h_j)$

where:
* $h_i = \sum_iW_jix_i + b_i$
* $x_i$ is input
* $W$ is weight
* $b$ is bias

```python
# Output (prediction) formula
def output_formula(features, weights, bias):
    # Wx is a dot
    return sigmoid(np.dot(features, weights) + bias)
```

Error function:

$Error(y, \hat y) = -ylog(\hat y) - (1 - y)log(1 - \hat y)$

```python
# Error (log-loss) formula
def error_formula(y, output):
    return - y*np.log(output) - (1 - y) * np.log(1-output)
```

The function that updates the weights (gradient descent step):

$w_i \rightarrow w_i + \alpha(y - \hat y)x_i$

$b \rightarrow b + \alpha(y - \hat y)$

where:
* $w_i$ is weight from set of weights $W$ (weights for each input)
* $\alpha$ is elarning rate
* $b$ is bias

```python
# Gradient descent step
def update_weights(x, y, weights, bias, learnrate):
    output = output_formula(x, weights, bias)
    d_error = -(y - output)
    weights -= learnrate * d_error * x
    bias -= learnrate * d_error
    return weights, bias
```

#### Training function
This function will help us iterate the gradient descent algorithm through all the data, for a number of epochs. It will also plot the data, and some of the boundary lines obtained as we run the algorithm.

```python
np.random.seed(44)

epochs = 100
learnrate = 0.01

def train(features, targets, epochs, learnrate, graph_lines=False):
    
    errors = []
    n_records, n_features = features.shape
    last_loss = None
    weights = np.random.normal(scale=1 / n_features**.5, size=n_features)
    bias = 0
    for e in range(epochs):
        del_w = np.zeros(weights.shape)
        for x, y in zip(features, targets):
            output = output_formula(x, weights, bias)
            error = error_formula(y, output)
            weights, bias = update_weights(x, y, weights, bias, learnrate)
        
        # Printing out the log-loss error on the training set        
        out = output_formula(features, weights, bias)
        loss = np.mean(error_formula(targets, out))
        errors.append(loss)
        if e % (epochs / 10) == 0:
            print("\n========== Epoch", e,"==========")
            if last_loss and last_loss < loss:
                print("Train loss: ", loss, "  WARNING - Loss Increasing")
            else:
                print("Train loss: ", loss)
            last_loss = loss
            predictions = out > 0.5
            accuracy = np.mean(predictions == targets)
            print("Accuracy: ", accuracy)
        if graph_lines and e % (epochs / 100) == 0:
            display(-weights[0]/weights[1], -bias/weights[1])
            

    # Plotting the solution boundary
    plt.title("Solution boundary")
    display(-weights[0]/weights[1], -bias/weights[1], 'black')

    # Plotting the data
    plot_points(features, targets)
    plt.show()

    # Plotting the error
    plt.title("Error Plot")
    plt.xlabel('Number of epochs')
    plt.ylabel('Error')
    plt.plot(errors)
    plt.show()
```

#### Time to train the algorithm!
When we run the function, we'll obtain the following:
* 10 updates with the current training loss and accuracy
* A plot of the data and some of the boundary lines obtained. The final one is in black. Notice how the lines get closer and closer to the best fit, as we go through more epochs.
* A plot of the error function. Notice how it decreases as we go through more epochs.

```python
train(X, y, epochs, learnrate, True)
```

## Perceptron vs Gradient Descent

The main difference is between the way how weights and bias are change:
* In perceptron: if $y - \hat y = 0$ then $w_i = w_i + \alpha x_i$ and if otherwise $w_i - \alpha x_i$
    * note: correctly classified: $y - \hat y = 0$
    * note: misclassified: $y - \hat y = 1$ or $y - \hat y = -1$
* In Gradient Descent: $w_i = w_i + \alpha(y - \hat y)x_i$

So the only difference is that Perceptron takes only values $-1, 0, 1$ and in Gradient Descent it takes the whole range between $-1$ and $1$!

In other words:
* in perceptron if classification is correct we do nothing.
* in Gradient Descent if:
    * point is misclassified he says "come closer" to the bounding line,
    * point is correctly classified he says "go away" to the bounding line.
